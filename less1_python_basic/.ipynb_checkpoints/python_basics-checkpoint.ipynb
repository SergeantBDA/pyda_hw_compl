## Задание 1

Даны 2 переменных, в которых хранятся строки произвольной длины: *phrase_1* и *phrase_2*.  
Напишите код, который проверяет какая из этих строк длиннее.

Примеры работы программы:  
1.  

```
phrase_1 = 'Насколько проще было бы писать программы, если бы не заказчики'
phrase_2 = '640Кб должно хватить для любых задач. Билл Гейтс (по легенде)'

```

Результат:  
`Фраза 1 длиннее фразы 2`
2.     

```
phrase_1 = '640Кб должно хватить для любых задач. Билл Гейтс (по легенде)'
phrase_2 = 'Насколько проще было бы писать программы, если бы не заказчики'
```

Результат:  
`Фраза 2 длиннее фразы 1`
3.     

```
phrase_1 = 'Насколько проще было бы писать программы, если бы не заказчики'
phrase_2 = 'Насколько проще было бы писать программы, если бы не заказчики'
```

Результат:  
`Фразы равной длины`

````python
phrase_1 = input("Введите фразу 1 и нажмите Enter - ")
phrase_2 = input("Введите фразу 2 и нажмите Enter - ")
len_phr1 = len(phrase_1)
len_phr2 = len(phrase_2)
if len_phr1 > len_phr2:
    rezult = "Фраза 1 длиннее фразы 2"
elif len_phr2 > len_phr1:
    rezult = "Фраза 2 длиннее фразы 1"
else:
    rezult = "Фразы равной длины"
print(rezult)
````

## Задание 2

Дана переменная, в которой хранится четырехзначное число (год). Необходимо написать программу, которая выведет, является ли данный год високосным или обычным.

Пример работы программы:
1.  

```
year = 2020
```
Результат:  
`Високосный год` 

2.  

```
year = 2019
```
Результат:  
`Обычный год` 

````python
year = int(input("Введите год - "))
"""
C 1582 считается если год кратен 4 и при этом не кратен 100 либо кратен 400. 
Год не является високосным, если он не кратен 4 либо кратен 4, 
но при этом кратен 100 и не кратен 400.
"""
if (
    (
        (year%4 == 0 and year%100 != 0) or 
        (year%400 == 0)
    ) and year > 1582
) or (year%4 == 0):
    rezult = "Високосный год"
else:
    rezult = "Обычный год"
print(rezult)
````

## Задание 3

Необходимо написать программу, которая будет запрашивать у пользователя месяц и дату рождения и выводить соответствующий знак зодиака.

Пример работы программы:  
1.  

```
Введите день:
30

Введите месяц:
Август
```
Результат:  
`Ваш знак зодиака: Дева` 

2.  

```
Введите день:
29

Введите месяц:
Октябрь
```
Результат:  
`Ваш знак зодиака: Скорпион` 


````python
import datetime as idt
mon = {"ЯНВАРЬ":1,"ФЕВРАЛЬ":2,"МАРТ":3,"АПРЕЛЬ":4,"МАЙ":5,"ИЮНЬ":6,
       "ИЮЛЬ":7,"АВГУСТ":8,"СЕНТЯБРЬ":9,"ОКТЯБРЬ":10,11:"НОЯБРЬ",12:"ДЕКАБРЬ"}
zodiak = {"Водолей":("2020.01.21", "2020.02.19"),
          "Рыбы":("2020.02.20", "2020.03.20"),
          "Овен":("2020.03.21", "2020.04.30"),
          "Телец":("2020.04.21", "2020.05.21"),
          "Близнецы":("2020.05.22", "2020.05.21"),
          "Рак":("2020.05.22", "2020.07.22"),
          "Лев":("2020.07.23", "2020.08.22"),
          "Дева":("2020.08.23", "2020.09.22"), 
          "Весы":("2020.09.23", "2020.10.22"),
          "Скорпион":("2020.10.23", "2020.11.22"),
          "Стрелец":("2020.11.22", "2020.12.21"), 
          "Козерог":("2020.12.22", "2021.01.20")}

def get_Z(dt, intrvl):
    rez = None
    for idx in intrvl:
        dt1 = idt.datetime.strptime(intrvl[idx][0], "%Y.%m.%d")
        dt2 = idt.datetime.strptime(intrvl[idx][1], "%Y.%m.%d")
        if dt1 <= dt <= dt2:
            rez = idx
    return(rez)
#В качестве названия месяца возможно ввести любую часть слова
#поиск будет осуществляться до первого совпадения
inp_mon = "Введите название месяца Вашего рождения - "
while True:
    mon_n = input(inp_mon)
    inp_mon = "Введите повторно НАЗВАНИЕ месяца Вашего рождения - "
    for idx in mon:
        find_mon = idx.rfind(mon_n.upper())
        if find_mon != -1:
            mon_n = idx
            break
    if find_mon != -1:
        break
    

dd = int(input("Введите число Вашего рождения - "))
#Проверка на ввод числа превышающих предельные значения дней месяца, 
#на точное определения зодика не влияет
if dd > 28:
    dd = 28
elif dd < 1:
    dd = 1
#Привязка знаков задиака к 2020 в соответствии с кортежем zodiak
yyyy = 2020
mm = mon[mon_n] 

#Проверка козерога, конец года 2020, начало 2021
if mm == 1 and dd < 20:
    yyyy = 2021
    
dt = idt.datetime(year = yyyy,
                   month = mm,
                   day = dd)
print(get_Z(dt, zodiak))
````

## Задание 4

Вам нужно написать программу для подбора упаковок по размерам товара. Размеры хранятся в переменных (в сантиметрах):

```
width = 10
length = 205
height = 5
```

Используйте следующие правила:
- если каждое из трех измерений менее 15 сантиметров, то выведите на экран "Коробка №1";
- если хотя бы одно из измерений больше 15 сантиметров, но менее 50 сантиметров, то выводите "Коробка №2";
- если длина товара больше 2 метров, то выводите "Упаковка для лыж";
- во всех остальных случаях выводите "Стандартная коробка №3".


````python
width = int(input("Ширина коробки (см) - " ))
length = int(input("Длина коробки (см) - " ))
height = int(input("Высота коробки (см) - " ))

if length > 200:
    print("Упаковка для лыж")
elif (width < 15) and (length < 15) and (height < 15):
    print("Коробка №1")
elif (15 < width < 50) or (15 < length < 50) or (15 < height < 50):
    print("Коробка №2")
else: 
    print("Стандартная коробка №3")
    
````

## Задание 5 (необязательное)

Дана переменная, в которой хранится шестизначное число (номер проездного билета). Напишите программу, которая будет определять, является ли данный билет "счастливым". Билет считается счастливым, если сумма первых трех цифр совпадает с суммой последних трех цифр номера.   

Примеры работы программы:  
1. ```number = 123456```

Результат:  
`Неасчастливый билет`
2. ```number = 123321```

Результат:  
`Счастливый билет`


````python
while True:
    number = input("Введите шестизначное число - ")
    try:
        len_num = len(number) 
        if (len_num == 6):
            num = int(number)
            break
    except Exception:
        pass
S1 = 0
S2 = 0
i  = 0
for sym in number:
    i  += 1
    if i <= len_num/2:
        S1 = S1 + int(sym)
    else:
        S2 = S2 + int(sym)

if S1 == S2:
    print("Счастливый билет")
else:
    print("Неасчастливый билет")
    
    
    
````
