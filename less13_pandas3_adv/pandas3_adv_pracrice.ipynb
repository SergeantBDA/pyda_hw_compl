###### Задание 1

Используйте файл с оценками фильмов ml-latest-small/ratings.csv. Посчитайте среднее время жизни пользователей, которые выставили более 100 оценок. Под временем жизни понимается разница между максимальным и минимальным значением столбца timestamp для данного значения userId.

```python
#Решение по заданию 1
import pandas as pd
RATINGS_PATH = "source/ml-latest-small/"

#загружаем файл
df_ratings = pd.read_csv(RATINGS_PATH + 'ratings.csv')

#опрделяем всех пользователей, которые поставили более 100 оценок
df_user_m100 = (df_ratings.groupby(by = 'userId').
                movieId.count().
                reset_index())

#переименовываем атрибуты фрейма
df_user_m100 = df_user_m100[(df_user_m100['movieId'] > 100)]
df_user_m100.columns = ['userId', 'count']

# операции: 
#1. группировка по пользователям, мининмальные и максимальные даты оценок 2
#2. отбираем из группировки #1 только пользователей, которые поставили свыше 100 оценок
# через объединение inner c df_user_m100
df_rezult = df_user_m100.merge( df_ratings.groupby(by = 'userId')['timestamp'].agg(['min', 'max']).reset_index(),
                                how = 'inner', 
                                on  = 'userId' )

#добавлем в результирующий фрейм время жизни
df_rezult['lifetime'] = pd.to_timedelta( (df_rezult['max'] - df_rezult['min']), unit = 's')

print('Результат:')
df_rezult[['userId', 'count', 'lifetime']]
```

Дана статистика услуг перевозок клиентов компании по типам (см. файл с кодом занятия).

Необходимо сформировать две таблицы:
- таблицу с тремя типами выручки для каждого client_id без указания адреса клиента
- аналогичную таблицу по типам выручки с указанием адреса клиента  

Обратите внимание, что в процессе объединения таблиц данные не должны теряться.

###### К Заданию 2
Дана статистика услуг перевозок клиентов компании по типам:
- rzd - железнодорожные перевозки
- auto - автомобильные перевозки
- air - воздушные перевозки
- client_base - адреса клиентов

```python
# Решение по заданию 2
# Исходные данные
rzd = pd.DataFrame({'client_id': [111, 112, 113, 114, 115],
                    'rzd_revenue': [1093, 2810, 10283, 5774, 981]})

auto = pd.DataFrame({'client_id': [113, 114, 115, 116, 117],
                     'auto_revenue': [57483, 83, 912, 4834, 98]})


air = pd.DataFrame({'client_id': [115, 116, 117, 118],
                    'air_revenue': [81, 4, 13, 173]})

client_base = pd.DataFrame({'client_id': [111, 112, 113, 114, 115, 116, 117, 118],
                            'address':   ['Комсомольская 4', 'Энтузиастов 8а', 'Левобережная 1а', 'Мира 14', 'ЗЖБИиДК 1', 
                                          'Строителей 18', 'Панфиловская 33', 'Мастеркова 4']})

# Функция добавляет новое поля с типом транспорта и
# одинаково переименовывает фреймы
def transf_df(df, columns_, str_): 
    df.columns = columns_
    df['transport'] = str_
    
    return df

#Объединение rzd, auto, air в один датафрейм с атрибутами ['client_id', 'revenue', 'transport']
df_trafic = pd.concat([transf_df(rzd,  ['client_id', 'revenue'], 'rzd'),
                         transf_df(auto, ['client_id', 'revenue'], 'auto'),
                         transf_df(air,  ['client_id', 'revenue'], 'air')]).sort_values(by = 'client_id').reset_index()

#Все клиенты
df_client   = df_trafic[['client_id']].drop_duplicates(subset = 'client_id')

#Типы транспорта, используемого клиентами (как я понял понятие по 3 типам выручки из задания)
df_tansport = df_trafic[['transport']].drop_duplicates(subset = 'transport')

#Декартово произведение клиентов с типами транпорта (все со всеми) используя поле с одинаковым занчением по всем строкам
df_tabl1 = df_client.assign(key = 0).merge(df_tansport.assign(key = 0), on = 'key').drop("key", axis=1)


df_tabl2 = df_tabl1.merge(client_base, how = 'left', on = 'client_id')

print('Результат:\n', df_tabl1, '\n'*2, df_tabl2)
```
