```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

#увеличим дефолтный размер графиков
from pylab import rcParams
rcParams['figure.figsize'] = 13,6
```

```python
# Решение

# В качестве анализируемого датафрейма используем данные по участникам олимпиад за 120 лет (1896-2016) с Kaggle 
# https://www.kaggle.com/heesoo37/120-years-of-olympic-history-athletes-and-results?select=athlete_events.csv
df = pd.read_csv('ath_events.csv', na_values = 'NaN')
df_region = pd.read_csv('noc_regions.csv')

df = df.merge(df_region[['NOC', 'region']], on = 'NOC', how = 'inner')

#Формируем датафремй для анализа: Год, Сезон, Пол, Количество участников
df_count_members_years = df.groupby(['Year','Season','Sex','ID']).size().reset_index().\
                            groupby(['Year','Sex','Season'])['ID'].count().reset_index()

#создаем полотно дл яотображения графика
figure, axes = plt.subplots()

#строим на полотне графики по количеству участников олимпиад, разделенных по гендарному признаку и сезонному виду спорта
for Season_ in df_count_members_years['Season'].unique():
    for Sex_ in df_count_members_years['Sex'].unique():
        df_tmp = df_count_members_years[((df_count_members_years['Season'] == Season_) & 
                                         (df_count_members_years['Sex'   ] == Sex_))]
        axes.plot(df_tmp.Year, df_tmp.ID, linewidth = 3, label = f'{Season_}, {Sex_}')

#форомируем разметку на осях
#основные линии оси-абсцис (Х) 
axes.xaxis.set_major_locator(ticker.MultipleLocator(5))
#вспомогательные линии оси-абсцис (Х)
axes.xaxis.set_minor_locator(ticker.MultipleLocator(1))
#основные линии оси-ординат (Х) 
axes.yaxis.set_major_locator(ticker.MultipleLocator(500))
#вспомогательные линии оси-ординат (Х) 
axes.yaxis.set_minor_locator(ticker.MultipleLocator(100))

#отображаем легенду
axes.legend(fontsize = 10, ncol = 2, facecolor = 'oldlace', edgecolor = 'r', title = 'Легенда', title_fontsize = '12'           )

#Подписываем оси
plt.title('Кол-во участников олимпийских игр по годам')
plt.xlabel('Год')
plt.ylabel('Кол-во участников') 

figure.set_figwidth(15)
figure.set_figheight(8)
axes.grid()
```

##### Гипотезы
1. Общие количество участников Олимпиад имеет тенденцию к увелечению, существующие провалы объясняются политической ситуацией в мире.
2. Количество атлетов мужского пола, принимающих участие в Олимпиаде, всегда было больше чем женщин, но за поледние 40 лет разрыв сущетсвенно сократился.
3. В Олимпиаде 1896 года не принимали участие женщины.
4. Зимние Олимпиды стали проводится позже летних игр на 28 лет.

```python
fields = ["ID", "Sex", "Age", "Height", "Weight", "Team", "NOC", "Games", "Year", "Season", "City"]
#Подготовка датафрейма для визуаллизации - анализ возраста атлетов, принимающих участие в Олимпийских играх
#1. ислючм из датафррейма вид спорта и медаль (1 участник может учавствовать в разных турнирах)
df_games = df[fields].drop_duplicates()
#2. отберем 20 самых масоовых по учатсию игр
df_games = df_games['Year'].value_counts().sort_values(ascending = False).head(20).reset_index().\
                              merge(df_games, how = 'inner', left_on = 'index', right_on = 'Year')
#3.построим для анализа сводную таблицу где индекс - ид.участника, колонки - год проведения, занчения - возраст участника
# визуализируем сводную таблицу в виде "ящиков с усами" (boxplot)
pd.pivot_table(df_games[['index', 'ID', 'Age']], values = 'Age', index = 'ID', columns = 'index').\
plot(kind='box',title='The age of participants of the Olympic games', yticks = range(10, 100, 5), ylim = (10,100), grid = True)
```

##### Гипотезы
1. Возраст атлетов, принимающих участие в Олимпийских играх от 20 до 30 лет
2. Самый возрастной (более 95 лет) участник был в 1928 году, самый молодой в 1988, 1992 гг. (11 лет).

```python
df_medal = pd.pivot_table(df.groupby(['region','Medal'])['ID'].count().reset_index(), 
                          values = 'ID', index = 'region', columns = 'Medal', aggfunc = 'sum').reset_index().fillna(0)

df_medal['Total'] = df_medal['Gold'] + df_medal['Silver'] + df_medal['Bronze']

df_medal = df_medal.sort_values('Total', ascending = False)

df_medal[['region', 'Total']].head(10).plot(kind = 'bar', 
                                           x = 'region',
                                           title = 'Медальный зачет за 120 летнию (1896 - 2016) историю Олимпийских игр',
                                           yticks = range(0, 6000, 500), ylim = (0,6000),
                                           grid = True)
```

##### Гипотезы
На первом месте в медальном зачете за 120-летнию историю Олимпийских игр со значительным перевесом находится США.
