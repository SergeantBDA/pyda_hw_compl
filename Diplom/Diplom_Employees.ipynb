##### 2. Рассчитайте основные статистики для переменных (среднее,медиана,мода,мин/макс,сред.отклонение). 

```python
# Задание 2. Вариант 1.

# StateBase - класс для получения основных статистических показателей датафрейма
# применяются ручные алгоритмы или встроенные алгоритмы датафрейма
# стоит обратить внимание на моду, в результат выводится 1-е значения моды, остальные игнорируются
# для получения всех занчений моды необходимо использовать метод - mmoda
class StateBase:
    
    def __init__(self, df_, numeric_columns_ = []):
        self.df = df_
        if len(numeric_columns_) == 0:
            #На этапе иницирования определяем чсиловые поля датафрейма
            self.numeric_columns = list(self.df.dtypes[(self.df.dtypes == 'int64'  ) | 
                                                       (self.df.dtypes == 'float64')].index)
        else:
            self.numeric_columns = numeric_columns_
            
    def mmean(self, column, precision = 16):
        """
        Ручное определение среднего значения серии
        """
        if column in self.numeric_columns:
            if len(self.df[column]) > 0:
                ret = round(sum(self.df[column])/len(self.df[column]), precision)
            else:
                ret = None
        else:
            ret = None
        return ret      

    def amean(self, column, precision = 16):
        """
        Определение среднего значения серии, метод датафрейма
        """
        if column in self.numeric_columns:
            ret = round(self.df[column].mean(), precision)
        else:
            ret = None
        return ret
    
    def mmedian(self, column, precision = 16):
        """
        Ручное определение медианы серии
        """
        mmedian_dict = {}
        if column in self.numeric_columns:
            list_ = sorted(df[column])
            count_list = len(list_)
            middle = count_list // 2
            if count_list%2 == 0:
                ret = round((list_[middle - 1] + list_[middle])/2, precision)
            else:
                ret = list_[middle]
        else:
            ret = None
        return ret

    def amedian(self, column, precision = 16):
        """
        Определение меданы серии, метод датафрейма
        """       
        if column in self.numeric_columns:
            ret = round(self.df[column].median(), precision)
        else:
            ret = None
        return ret
    
    def mmoda(self, column):
        """
        Ручное определение моды серии.
        Возвращет словарь {Значение: Количество повторений}
        """
        if column in self.numeric_columns:
            moda_dict = {}
            values = list(self.df[column].unique())
            moda_dict = {val: len(self.df[df[column] == val]) for val in values}
            max_val = max( moda_dict.values() )
            ret = None 
            #если все занчения встречаются 1 раз, то выборка моды не имеет
            if max_val > 1:
                ret = {val: cnt for val, cnt in moda_dict.items() if cnt == max_val}          
        else:
            ret = None
        return ret

    def amoda(self, column):
        """
        Определение моды серии, метод датафрейма
        """
        if column in self.numeric_columns:
            ret = self.df[column].mode()[0]
        else:
            ret = None
        return ret
    
    def amin(self, column):
        """
        Определение минимума серии, метод датафрейма
        """
        if column in self.numeric_columns:
            ret = self.df[column].min()
        else:
            ret = None
        return ret
    
    def amax(self, column):
        """
        Определение максимума серии, метод датафрейма
        """
        if column in self.numeric_columns:
            ret = self.df[column].max()
        else:
            ret = None
        return ret
    
    def mstd(self, column, precision = 16):
        """
        Русное определение стандартного отклонения серии, метод датафрейма
        """
        mean_ = self.amean(column, 16)
        summa = 0
        cnt = len(self.df[column])
        if cnt > 1:
            for val in self.df[column]:
                summa += (mean_ - val)**2
            ret = round( (summa/(cnt - 1))**0.5, precision )
        else:
            ret = None
        return ret
 
    def astd(self, column, precision = 16):
        """
        Определение стандартного отклонения серии, метод датафрейма
        """
        if column in self.numeric_columns:
            ret = round( self.df[column].std(), precision)
        else:
            ret = None
        return ret
    
    def acnt(self, column):
        """
        Определение количество элементов, метод датафрейма
        """
        if column in self.numeric_columns:
            ret = self.df[column].count()
        else:
            ret = None
        return ret
    
    # Методы для получения статистических знaчений по всем чиловым полям
    # precision - точность округления после запятой, flag - признак ручного расчета или преименения метода датафрейма
    def get_means(self, precision = 16, flag = 'auto'):
        """
        Получение среднего по всем полям
        """     
        mmean_dict = {}
        for column in self.numeric_columns:
            if flag != 'auto':
                mmean_dict[column] = self.mmean(column, precision)
            else:
                mmean_dict[column] = self.amean(column, precision)
        return mmean_dict
    
    def get_medians(self, precision = 16, flag = 'auto'):
        """
        Получение медианы по всем полям
        """
        mmedian_dict = {}
        for column in self.numeric_columns:
            if flag == 'auto':
                mmedian_dict[column] = self.amedian(column, precision)
            else:
                mmedian_dict[column] = self.mmedian(column, precision)
        return mmedian_dict
    
    def get_mods(self, precision = 16, flag = 'auto'):
        """
        Получение моды по всем полям
        """
        mmods_dict = {}
        for column in self.numeric_columns:
            if flag == 'auto':
                mmods_dict[column] = self.amoda(column)
            else:
                mmods_dict[column] = list(self.mmoda(column))[0]
        return mmods_dict
    
    def get_mins(self, precision = 16, flag = 'auto'):
        """
        Получение минимума по всем полям
        """
        amin_dict = {}
        for column in self.numeric_columns:
            amin_dict[column] = self.amin(column)
        return amin_dict

    def get_maxs(self, precision = 16, flag = 'auto'):
        """
        Получение максимума по всем полям
        """
        amax_dict = {}
        for column in self.numeric_columns:
            amax_dict[column] = self.amax(column)
        return amax_dict
    
    def get_stds(self, precision = 16, flag = 'auto'):
        """
        Получение среднего отклонения по всем полям
        """
        std_dict = {}
        for column in self.numeric_columns:
            if flag == 'auto':
                std_dict[column] = self.astd(column, precision)
            else:
                std_dict[column] = self.mstd(column, precision)
        return std_dict
    
    def get_cnt(self, precision = 16, flag = 'auto'):
        """
        Получение количество значений по всем полям
        """
        acnt_dict = {}
        for column in self.numeric_columns:
            acnt_dict[column] = self.acnt(column)
        return acnt_dict   
    
    FUNC = {'Кол-во'  :get_cnt,
            'Среднее' :get_means,
            'Стд.откл':get_stds,
            'Мода'    :get_mods,
            'Миниум'  :get_mins,
            'Медиана' :get_medians,
            'Максимум':get_maxs}
    
    #Метод для формирования итогового датафрейма, используется атрибут FUNC
    def main(self, precision = 16, flag = 'auto'):
        
        df_ = pd.DataFrame()
        for key_, func_ in self.FUNC.items():
            df_ = df_.append( pd.DataFrame( func_(self, precision, flag), index = [key_] ) )
        return(df_)

st = StateBase(df)
st.main(16, 'auto')
```

##### 3. Рассчитайте и визуализировать корреляционную матрицу для количественных переменных.
##### Определите две самые скоррелированные и две наименее скоррелированные переменные.


```python
# Задание 3.
def get_corr(df_, cnt_ = 2, method_ = 'pearson'):
    """
    Функция возвращает попарный список сочетаний атрибутов и значения корреляции по датафрейму df_,
    второй параметор количество наиболее и наименнее скоррелированных переменных 
    третий параметор метод корреляции {'pearson', 'kendall', 'spearman'}
    """
    #Формируем таблицу корреляции
    df_corr   = df_.corr(method = method_)
    
    #Формруем список всех сочетаний полей таблицы корреляции
    list_comb = list(ite.combinations(df_corr.columns, 2))
    
    #Пустой датафрейм
    df_ret = pd.DataFrame()
    
    #Цикл создания датафрейма с парами атрибутов и значениями корреляиций
    for item in list_comb:
        list_ = list(item)
        list_.append(df_corr.at[item])
        #модуль значения корреляции для последующей сортировки от 0 до 1
        list_.append(abs(df_corr.at[item]))
        list_.append(method_)
        
        #Формируем строку фрейма
        df_ret = df_ret.append( pd.DataFrame([list_], 
                                             columns = ['Attr1', 'Attr2', 'Corr', 'Corr_abs', 'Method']), 
                                             ignore_index = True ) 
    
    #Сортируем фрейм по взрастанию убыванию модуля корреляции
    df_ret = df_ret.sort_values(by=['Corr_abs'],ascending=False)[['Attr1', 'Attr2', 'Corr', 'Method']]
    
    #Отбираем строки отсортированного списка с начала и конца
    if cnt_ < len(df_ret) // 2:
        df_ret = pd.concat([df_ret[:cnt_], 
                            df_ret[-cnt_:]])

    return df_ret

#Разделительная строка методов расчета корреляции
df_none = pd.DataFrame([['.', '.', None, '.']], columns = ['Attr1', 'Attr2', 'Corr',  'Method'])

#Формируем итоговый датафрейм разными методами
pd.concat( [get_corr(df, 2), df_none,
            get_corr(df, 2, 'kendall'), df_none,
            get_corr(df, 2, 'spearman')] )

```

*Атрибуты наиболее сокррелированных значений по всем методам сходятся, наименее скоррелированные значения по методу Пирсона отличаются.*

##### 8. Рассчитать следующие показатели среди уволившихся и не уволившихся сотрудников (по отдельности):
- Доля сотрудников с повышением за последние 5 лет
- Cредняя степень удовлетворенности
- Среднее количество проектов

```python
#Задание 8 (1) (Доля сотрудников с повышением за последние 5 лет)
df_promot = pd.pivot_table(df, 
                           index   = 'left', 
                           columns = 'promotion_last_5years', 
                           aggfunc = 'count', 
                           values  = 'salary').reset_index()

df_promot['portion'] = round(df_promot[1]/(df_promot[0] + df_promot[1]), 3)

left_dict = {0:'Среди не уволившихся', 1:'Среди уволившихся'}

for key, val in df_promot.iterrows():
    print( f'{left_dict[key]} {(val[0] + val[1]):5.0f} чел. было повышено в должности (за 5 лет) {val[1]:4.0f} чел., \
что составляет {val["portion"]:3.1%}')
```

*Повышение в должности мотивирует на сохранение работы*

```python
#Задание 8 (2) (Cредняя степень удовлетворенности)
#Группировка со сбросом индекса по left с получением среднего по satisfaction_level, округленного до 2 знака 
df_stsf_lev = df.groupby('left')['satisfaction_level'].mean().round(2).reset_index()

#Создание дополнительного датафрейма для расшифровки результата с применением list comprehension
df_left_sl = pd.DataFrame([[key, 'Степень удовлетворенности ' + val.lower()] for key, val in left_dict.items()], 
                        columns = ['id', 'comment'])

#Объединение датафремов для наглядного отображения результата
df_stsf_lev.merge(df_left_sl, how = 'inner', left_on = 'left', right_on = 'id')[['comment', 'satisfaction_level']]

```

*Увольняются менее удовлетворенные работники*

```python
#Задание 8 (3) (Среднее количество проектов)
#Принцип аналогичный Заданию 8 (2)
df_num_proj = df.groupby('left')['number_project'].mean().round(2).reset_index()

df_left_np = pd.DataFrame([[key, 'Среднее количество проектов ' + val.lower()] for key, val in left_dict.items()], 
                          columns = ['id', 'comment'])

df_num_proj.merge(df_left_np, how = 'inner', left_on = 'left', right_on = 'id')[['comment', 'number_project']]
```

##### 9. Разделить данные на тестовую и обучающую выборки. Построить модель LDA, предсказывающую уволился ли сотрудник на основе имеющихся факторов (кроме department и salary). Оценить качество модели на тестовой выборки.

```python
# Задание 9 (1).

# Разделить данные на тестовую и обучающую выборки
from sklearn.model_selection import train_test_split

attr_list = ['satisfaction_level','last_evaluation','number_project','average_montly_hours',
             'time_spend_company','Work_accident','promotion_last_5years']

X = df[attr_list]
Y = df['left']
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3)

print(f'Тестовых данных: {len(X_test)}, обучающих: {len(X_train)}, тестовых: {len(X_test)/(len(X_test) + len(X_train)):3.1%} ' )
```

```python
# Задание 9 (2).
# Построить модель LDA, предсказывающую уволился ли сотрудник на основе имеющихся факторов (кроме department и salary).

# импорт LDA-функционала
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
# создание lda-объекта 
lda = LinearDiscriminantAnalysis()

# обучение модели
lda.fit( X_train, Y_train)
```

```python
# Задание 9 (3).
# Оценить качество модели на тестовой выборки.
from sklearn.metrics import accuracy_score
# сравнение факта и прогноза
# получаем прогнозное значение на основе X_test
Y_predict = lda.predict(X_test)
print(f'Оценка качества модели метод "accuracy_score": {accuracy_score(Y_test, Y_predict):4.2} ')

#функция для ручной оценки качества результата работы модели
def cmp_val(row):
    ret = 1
    if (row['predict'] != row['fact']):
        ret = 0
    return ret

#Ручной расчет качества модели
#датафрейм с тетовыми и смоделированными результатами
df_compare = pd.DataFrame([list(Y_test), list(Y_predict)]).T
df_compare.rename({0:'fact', 1:'predict'}, axis = 1, inplace = True)

#добавляем поле 'compare' с результатом среавнения Y_test, Y_predict (0 - значения расходятся, 1 - прогноз совпал)
df_compare['compare'] = df_compare.apply(cmp_val, axis = 1)

#группируем по полю 'compare' и находим количество значений (0 - значения расходятся, 1 - прогноз совпал)
df_compare_total = df_compare.groupby('compare')['fact'].count().reset_index()

test_len   = df_compare_total['fact'].sum()
test_equal = df_compare_total[(df_compare_total['compare'] == 1)]['fact'][1]
print(f'\nРучная оценка качества прогноза:\n\
Общее количество тестовых данных {test_len} совпадения {test_equal} сходимость прогноза: {(test_equal/test_len):3.1%}'  ,  )
```
