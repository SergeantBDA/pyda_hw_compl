- объем выборки n = 25;
- среднее значение по выборке $\overline x$ = 130;
- Из достоверных источников известно стандартное отклонение $\sigma = 12$.

##### Решение по заданию 1

```python
class cond_int:
    """
    Класс для построения доверительного интервала с заданными параметрами
    """    
    
    def __init__(self, x_avg_, sigma_, n_, b_):
        """
        x_avg_ - среднее
        sigma_ - cтандартное отклонение
        n_     - объем выборки
        b_   - доверительный интервал [0:1] для вычисления квантиля нормального распределения по словарю QUANTILES
        """
        self.x_avg = x_avg_
        self.sigma = sigma_
        if n_ > 0: self.n = n_
        self.Z_a   = stats.t.ppf(1 - (1 - b_)/ 2, n_) 
        # https://www.medcalc.org/manual/t-distribution.php
        # https://ru.stackoverflow.com/questions/917453/%D0%9A%D0%B0%D0%BA-%D0%BF%D0%BE%D0%BB%D1%83%D1%87%D0%B8%D1%82%D1%8C-%D0%B7%D0%BD%D0%B0%D1%87%D0%B5%D0%BD%D0%B8%D0%B5-%D1%82-%D1%81%D1%82%D0%B0%D1%82%D0%B8%D1%81%D1%82%D0%B8%D0%BA%D0%B8
    
    def delta(self, accuracy_ = 16):
        """
        Возвращает дельту для расчета доверительного интвревала
        """
        return round(self.Z_a * ( (self.sigma) / (self.n)**0.5 ), accuracy_)
    
    def x_beg(self, accuracy_ = 16):
        """
        Возвращает начальное значение доверительного интервала
        """        
        return( self.x_avg - self.delta(accuracy_) )
    
    def x_end(self, accuracy_ = 16):
        """
        Возвращает конечное значение доверительного интервала
        """       
        return( self.x_avg + self.delta(accuracy_) )
    
    def main(self):
        """
        Возвращает доверительный интревал
        """ 
        return (self.x_beg(), self.x_avg, self.x_end())


interval = cond_int(130, 12, 25, 0.98)
interval.main()
```

##### Решение по заданию 2

```python

population_men   = stats.norm.rvs(loc = 171, scale = 19, size = 11000000)  # Выборка мужчин со средним ростом 171
population_women = stats.norm.rvs(loc = 165, scale = 16, size = 12000)  # Выборка женщин со средним ростом 165

### Ваш код...

print( stats.ttest_ind(population_men, population_women) )

print( stats.ttest_1samp(a=population_women, popmean=population_men.mean()) )



#Визуализация
fig, ax = plt.subplots()
ax.hist(population_men,   bins=100);
ax.hist(population_women, bins=100);
fig.set_figwidth(15)
fig.set_figheight(10)
```

##### Решение по заданию 3

```python
QUANTILES = {0.9   :1.645, 
             0.95  :1.96, 
             0.975 :2.241, 
             0.98  :2.326, 
             0.9   :2.576, 
             0.995 :2.807, 
             0.999 :3.291, 
             0.9995:3.481,
             0.9999:3.891}
    #Источник https://studfile.net/preview/3795082/page:2/
    
def get_n(q_, sigma_, delta_):
    """
    Функция для получения объема выборки
    """
    return ( (q_ * sigma_)/delta_ )**2
    
print (f'Объем необходимой выборки для исследования среднего чека за кофе: {get_n(interval.QUANTILES[0.95], 150, 50 )}')
print ('-------------------------------------------------------------------')
```

Представьте, что вы хотите разоблачить "волшебника" который считает, что умеет предсказывать погоду на завтра, отвечая просто: дождь или солнце. Вы пронаблюдали за ответами "волшебника" в течении какого-то периода времени и получили некие результаты (см.ниже).

Можно ли сказать, что маг действительно умеет предсказывать погоду? Уровень значимости принять за `0.05` ($\alpha=0.05$).
</div>

```python
observations = pd.DataFrame( [[10,15],[39,34]],
                             index=['Дождь','Солнце'],
                             columns=['Ответ волшебника','Реальность'])
observations
```

##### Решение по заданию 4

```python
alfa = 0.05
total_predictions = len(observations)
errors            = abs(observations.iloc[0, 0] - observations.iloc[0, 1]) + \
                    abs(observations.iloc[1, 0] - observations.iloc[1, 1])

erroros_per       = errors/total_predictions

if erroros_per < alfa:
    print('Волшебник умеет предсказывать погоду!')
else:
    print('Волшебник НЕ умеет предсказывать погоду!')
print('----------------------------------------')
```
